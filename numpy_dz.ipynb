```python
# 36 Как вычислить средние значения, используя скользящее окно по массиву?
a = np.arange(20) + 1
ws = 6
cms_a = np.cumsum(a)
cms_a[ws:] = cms_a[ws:] - cms_a[:-ws]
cms_a = cms_a[ws - 1:] / ws
print(f'Исходный вектор:\n{a}\nРазмер окна:\n{ws}\nСредние по окну:\n{cms_a}')
```

```python
# 37 Рассмотрим одномерный массив Z, постройте из него двумерный массив,
# первая строка которого (Z[0],Z[1],Z[2]), а каждая последующая строка
# сдвинута на 1 (последняя строка должна быть (Z[ -3],Z[-2],Z[-1])
n = 3
a = np.random.randint(1, 11, 10)
sh = (a.size - n + 1, n)
st = (a.itemsize, a.itemsize)
b = np.lib.stride_tricks.as_strided(a, shape=sh, strides=st)
print(f'Исходный:\n{a}\nПреобразованный:\n{b}')
```

```python
# 38 Рассмотрим 2 набора точек P0,P1, описывающих линии (2d), и набор точек P.
# Как вычислить расстояние от каждой точки j (P[j]) до каждой линии i (P0[i],P1[i])?
n = int(input('Введите количество прямых: '))
m = int(input('Введите количество точек: '))
a = np.random.randint(0, 11, (n, 2))
b = np.random.randint(0, 11, (n, 2))
p = np.random.randint(0, 11, (m, 2))
print(f'Массив начальных точек прямых:\n{a}\nМассив конечных точек прямых:\n{b}\nМассив точек p:\n{p}')
for i in range(m):
    for j in range(n):
        w = np.cross((b[j]-a[j]), (p[i]-a[j]))
        s = np.linalg.norm(w)
        ab = np.linalg.norm(b[j]-a[j])
        d = s / ab
        print(f'Расстояние от точки {i+1} до прямой {j+1} = {d}')
```
